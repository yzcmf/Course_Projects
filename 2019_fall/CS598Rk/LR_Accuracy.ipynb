```python
# import binascii
# def hash1(s): return int(binascii.hexlify(s), 16)
# def hash2(s): return ''.join(format(ord(i), 'b') for i in s)
import hashlib
d1 = {}
def md5_1(s): 
    h = int(hashlib.md5(s.encode()).hexdigest(),16)
    d1[h] = s  
    return h
d2 = {}
def md5_2(s): 
    h = int(hashlib.md5(s.encode()).hexdigest(),16)
    d2[h] = s  
    return h 
```
